### Helper Function

```python
def plot_image(image):
    plt.imshow(image, cmap="gray", interpolation="nearest")
    plt.axis("off")
```

### Load Data

```python
plt.figure(figsize=(20,20))

folder = "./pak_rup_data"

for i in range(5):
    file = random.choice(os.listdir(folder))
    folderPath = os.path.join(folder, file)
    randImg = random.choice(os.listdir(folderPath))
    imgPath = os.path.join(folderPath, randImg)
    img = mpimg.imread(imgPath)
    ax = plt.subplot(1, 5, i+1)
    ax.title.set_text(file)
    plot_image(img)
```

### Extract Labels, Resize Images, and Filter

```python
X_full = []
class_name = []

for group in os.listdir(folder):
    for img in os.listdir(os.path.join(folder, group))[:463]: # Only take 463 from each category to have the same number of samples
        imgPath = os.path.join(folder, group, img)
        if imgPath.lower().endswith(('.png', '.jpg', '.jpeg')): # Filter out non-standard img types such as '.gif'
            img = cv2.imread(imgPath)
            img = cv2.resize(img, (256, 256), interpolation = cv2.INTER_AREA) # Resize to uniform compact size
            img = np.array(img)
            img = img/255
            
            X_full.append(img)
            class_name.append(group)
```

```python
target_dict = {k : v for v, k in enumerate(np.unique(class_name))}
target_dict
```

```python
def get_key(val):
    for key, value in target_dict.items():
         if val == value:
                return key
 
    return "key doesn't exist"
```

```python
y_full = [target_dict[class_name[i]] for i in range(len(class_name))]
```

```python
y_full
```

### Split Data

```python
X_train, X_complement, y_train, y_complement = train_test_split(X_full, y_full, train_size=.5)
X_valid, X_test, y_valid, y_test = train_test_split(X_complement, y_complement, train_size=.5)
```

```python
X_train = np.array(X_train)
X_valid = np.array(X_valid)
X_test = np.array(X_test)

y_train = np.array(y_train)
y_valid = np.array(y_valid)
y_test = np.array(y_test)
```

```python
y_test_values = []

for i in y_test:
    y_test_values.append(get_key(i))
```

```python
y_test_values
```

```python
from functools import partial

DefaultConv2D = partial(tf.keras.layers.Conv2D,
                        kernel_size = 3, strides = 2, activation = 'relu', padding = "SAME")

cnn = tf.keras.models.Sequential([
    DefaultConv2D(filters = 64, input_shape = [256, 256, 3]),
    tf.keras.layers.MaxPooling2D(pool_size = 2),
    DefaultConv2D(filters = 128),
    tf.keras.layers.MaxPooling2D(pool_size = 2),
    DefaultConv2D(filters = 256),
    tf.keras.layers.MaxPooling2D(pool_size = 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units = 7, activation = 'softmax'),
])

cnn.compile(loss = "sparse_categorical_crossentropy", 
            optimizer = tf.keras.optimizers.Adam(), metrics = ["accuracy"])
```

```python
cnn.summary()
```

```python
history = cnn.fit(X_train, y_train, epochs=30, validation_data=(X_valid, y_valid))
```

```python
cnn.evaluate(X_test, y_test, verbose=0)
```

```python
loss_per_epoch = history.history['loss']
plt.plot(range(len(loss_per_epoch)), loss_per_epoch)
plt.title('Loss Over Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
testIn = 0

plot_image(X_test[testIn])
index = cnn.predict(np.expand_dims(X_test[testIn], axis=0))
print('pred:', get_key(np.argmax(index)))
print('ans:', get_key(y_test[testIn]))
```

```python
pred = []

rawPred = cnn.predict(X_test)

for i in rawPred:
    pred.append(get_key(np.argmax(i)))
```

```python
matrix = confusion_matrix(y_test_values, pred, labels=['10','100', '1000', '20', '50', '500', '5000'])
matrixDF = pd.DataFrame(matrix, index = ['10','100', '1000', '20', '50', '500', '5000'], columns = ['10','100', '1000', '20', '50', '500', '5000'])

plt.figure(figsize=(5,4))
sns.heatmap(matrixDF, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
total = 0
correct = 0

for i in range(len(pred)):
    total +=1
    if pred[i] == y_test_values[i]:
        correct += 1

# Accuracy        
correct / total
```
